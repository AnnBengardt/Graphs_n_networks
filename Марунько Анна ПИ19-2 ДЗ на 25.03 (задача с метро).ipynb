```python
warehouses = []
nodes = list(metro.nodes())

def calculate(graph, nodes, warehouse):
    excluded_nodes = []
    for node in nodes:
        if warehouse != node:
            if nx.astar_path_length(graph, warehouse, node) <= 25:
                excluded_nodes.append(node)
    return excluded_nodes


for node in nodes:
    warehouses.append(node)
    warehouses=warehouses[::-1]
    excluded = calculate(metro, nodes, node)
    for nd in excluded:
        nodes.remove(nd)
        
print("Необохдимо поставить склады на станциях: ")
print(warehouses)
```
